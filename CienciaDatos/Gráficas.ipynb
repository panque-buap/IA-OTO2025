```python
#Este código muestra el uso de la biblioteca numpy

# 1. Importar la librería NumPy
import numpy as np

# 2. Crear un array a partir de una lista
lista_numeros = [1, 2, 3, 4, 5]
mi_array = np.array(lista_numeros)
print(f"Array creado desde una lista: {mi_array}")

# 3. Crear un array con un rango de números
array_rango = np.arange(0, 10, 1) # Desde 0 hasta 9, de 2 en 2
print(f"Array con un rango: {array_rango}")

# 4. Realizar operaciones aritméticas elemento a elemento
array_multiplicado = mi_array * 2
print(f"Array multiplicado por 2: {array_multiplicado}")

array_sumado = mi_array + 5
print(f"Array sumado por 5: {array_sumado}")

# 5. Acceder a elementos del array (indexación)
elemento_dos = mi_array[1] # Accede al segundo elemento (índice 1)
print(f"Segundo elemento del array: {elemento_dos}")

# 6. Crear un array multidimensional
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
print(f"Array 2D:\n{array_2d}")

# 7. Acceder a un elemento en un array 2D
elemento_1_2 = array_2d[0, 1] # Fila 0, columna 1
print(f"Elemento en la primera fila, segunda columna: {elemento_1_2}")

# 8. Realizar operaciones con arrays 2D
suma_filas = array_2d + 10
print(f"Array 2D sumado por 10:\n{suma_filas}")
```

```python
#Este código muestra como realizar un gráfico de barras horizontales 
#utilizando matplotlib

import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['Categoría A', 'Categoría B', 'Categoría C']
valores = [20, 35, 30]

# Crear el gráfico de barras horizontal
plt.barh(categorias, valores, color='skyblue')

# Añadir título y etiquetas a los ejes
plt.title('Gráfico de Barras Horizontal de Ejemplo')
plt.xlabel('Valores')
plt.ylabel('Categorías')

# Mostrar el gráfico
plt.show()
```

```python
#Este código muestra como realizar un gráfico de pastel
#utilizando matplotlib

import matplotlib.pyplot as plt

# Datos de ejemplo
tamaños = [30, 25, 15, 10, 20]
etiquetas = ['Manzanas', 'Naranjas', 'Plátanos', 'Uvas', 'Mangos']
colores = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'orange']
# Para explotar un sector, crea una lista de ceros con un valor para el sector que quieres "sacar"

# Crear el gráfico de pastel
plt.pie(tamaños,
        labels=etiquetas,
        colors=colores,
        autopct='%1.1f%%') # Muestra los porcentaje



# Mostrar el gráfico
plt.title("Gráfico de Pastel de Frutas")
plt.show()
```

```python
#Este código muestra como realizar un histograma con datos aleatorios que siguen una distribución normal
#utilizando matplotlib y numpy

import numpy as np
import matplotlib.pyplot as plt

# 1. Generar datos aleatorios con distribución normal
media = 0  # Media (pico de la campana)
desviacion_estandar = 1  # Desviación estándar (plano de la distribución)
num_muestras = 1000  # Número de puntos de datos a generar

datos_normales = np.random.normal(loc=media, scale=desviacion_estandar, size=num_muestras)

# 2. Crear el gráfico (histograma)
plt.figure(figsize=(8, 6))  # Tamaño de la figura del gráfico
plt.hist(datos_normales, bins=30, color='g', label='Distribución Normal') # Crea un histograma

# 3. Añadir títulos y etiquetas
plt.title('Gráfico de Distribución Normal (Gaussiana)')
plt.xlabel('Valores')
plt.ylabel('Densidad de Frecuencia')
plt.grid(True)
plt.legend()

# 4. Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Crear una figura y una rejilla de 2 filas y 2 columnas de subgráficos
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
categorias = ['A', 'B', 'C', 'D']
valores_barras = [10, 20, 15, 25]

# 2. Dibujar en cada eje (subgráfico)
# Gráfico 1: Gráfico de líneas en el primer subgráfico (arriba-izquierda)
axes[0, 0].plot(x, y1, label='Seno')
axes[0, 0].set_title('Gráfico de Líneas')
axes[0, 0].legend()

# Gráfico 2: Otro gráfico de líneas en el segundo subgráfico (arriba-derecha)
axes[0, 1].plot(x, y2, color='red', label='Coseno')
axes[0, 1].set_title('Otro Gráfico de Líneas')
axes[0, 1].legend()

# Gráfico 3: Gráfico de barras en el tercer subgráfico (abajo-izquierda)
axes[1, 0].bar(categorias, valores_barras, label='Barras')
axes[1, 0].set_title('Gráfico de Barras')
axes[1, 0].legend()

# Gráfico 4: Un área vacía o un gráfico diferente en el cuarto subgráfico (abajo-derecha)
axes[1, 1].text(0.5, 0.5, 'Cuarto Subgráfico', ha='center', va='center')
axes[1, 1].set_title('Texto')

plt.savefig("cuatroGraficos.jpg")

# 3. Ajustar el diseño y mostrar el gráfico
plt.show()
```
